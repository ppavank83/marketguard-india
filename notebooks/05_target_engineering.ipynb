### target idea:

At date T, using only information available at T,
what happened to the stock after T?

### path to files and load them

```python
def find_project_root(start_path: Path) -> Path:
    """
    Walk upward from the current notebook folder until we find the project root.
    Project root is identified by config.yaml or README.md.
    """
    start_path = start_path.resolve()

    for path in [start_path] + list(start_path.parents):
        if (path / "config.yaml").exists() or (path / "README.md").exists():
            return path

    raise FileNotFoundError("Project root not found. Expected config.yaml or README.md.")


PROJECT_ROOT = find_project_root(Path.cwd())

INTERIM_DIR = PROJECT_ROOT / "data" / "interim"
FEATURE_DIR = INTERIM_DIR / "features"
TARGET_DIR = INTERIM_DIR / "targets"

TARGET_DIR.mkdir(parents=True, exist_ok=True)

FEATURE_INPUT_PARQUET = FEATURE_DIR / "stock_features_v1.parquet"
FEATURE_INPUT_CSV = FEATURE_DIR / "stock_features_v1.csv"

TARGET_OUTPUT_PARQUET = TARGET_DIR / "stock_features_with_targets_v1.parquet"
TARGET_OUTPUT_CSV = TARGET_DIR / "stock_features_with_targets_v1.csv"

TARGET_QUALITY_PATH = TARGET_DIR / "target_quality_summary_v1.csv"
TARGET_DICTIONARY_PATH = TARGET_DIR / "target_dictionary_v1.csv"

print("Project root:", PROJECT_ROOT)
print("Feature parquet:", FEATURE_INPUT_PARQUET)
print("Target dir:", TARGET_DIR)

if FEATURE_INPUT_PARQUET.exists():
    stock_features = pd.read_parquet(FEATURE_INPUT_PARQUET)
    print("Loaded from parquet")
else:
    stock_features = pd.read_csv(FEATURE_INPUT_CSV, parse_dates=["date"])
    print("Loaded from CSV")

stock_features["date"] = pd.to_datetime(stock_features["date"])

stock_features = (
    stock_features
    .sort_values(["yf_ticker", "date"])
    .reset_index(drop=True)
)
# to display all the columns in the DataFrame without truncation
pd.set_option("display.max_columns", None)
print("Stock features shape:", stock_features.shape)
print("Date range:", stock_features["date"].min(), "to", stock_features["date"].max())
print("Unique stocks:", stock_features["yf_ticker"].nunique())

display(stock_features.sample(5))
```

### Target engineering config and source-column check

For targets, we will use adj_close, not raw close, because it is adjusted for corporate actions.

```python
# Target horizons:
# 5 trading days  ≈ 1 week
# 20 trading days ≈ 1 month
# 60 trading days ≈ 3 months
TARGET_HORIZONS = [5, 20, 60]

# Main horizon for v1 model/risk labels
MAIN_TARGET_HORIZON = 20

# Downside / upside thresholds for classification-style labels
# These are not trading rules yet; they are target labels for analysis/model training.
DOWNSIDE_RISK_THRESHOLD_20D = -0.05   # -5% or worse over next 20 trading days
BIG_DOWNSIDE_THRESHOLD_20D = -0.10    # -10% or worse over next 20 trading days
UPSIDE_THRESHOLD_20D = 0.05           # +5% or better over next 20 trading days

required_source_cols = [
    "date",
    "symbol",
    "yf_ticker",
    "adj_close",
    "close",
    "model_ready_v1",
    "model_ready_v1_clean",
    "stock_acceptance_status",
    "manual_review_large_move_flag",
]

missing_source_cols = [
    col for col in required_source_cols
    if col not in stock_features.columns
]

print("Missing source columns:", missing_source_cols)

if missing_source_cols:
    raise ValueError(f"Missing source columns: {missing_source_cols}")

print("Target horizons:", TARGET_HORIZONS)
print("Main target horizon:", MAIN_TARGET_HORIZON)

print("\nAdj close missing values:")
print(stock_features["adj_close"].isna().sum())

print("\nModel-ready rows available before target creation:")
display(stock_features["model_ready_v1"].value_counts())

display(
    stock_features[
        [
            "date",
            "symbol",
            "yf_ticker",
            "adj_close",
            "close",
            "model_ready_v1",
            "model_ready_v1_clean",
            "stock_acceptance_status",
        ]
    ]
    .sample(20)
)
```

### Future return targets

Now create the basic future-return targets.

#### From today, what was the stock return after 5 / 20 / 60 trading days?

```python
stock_targets = stock_features.copy()

# Create future adjusted-close prices and future returns.
# Example:
# future_return_20d at date T =
#     adjusted close 20 trading days after T / adjusted close at T - 1
for horizon in TARGET_HORIZONS:
    future_price_col = f"future_adj_close_{horizon}d"
    future_return_col = f"future_return_{horizon}d"
    target_available_col = f"target_available_{horizon}d"

    stock_targets[future_price_col] = (
        stock_targets
        .groupby("yf_ticker")["adj_close"]
        .shift(-horizon)
    )

    stock_targets[future_return_col] = (
        stock_targets[future_price_col] / stock_targets["adj_close"] - 1
    )

    # 1 means the future target is available.
    # Last N rows per stock will be 0 because we do not know the future after dataset end.
    stock_targets[target_available_col] = (
        stock_targets[future_return_col].notna()
    ).astype(int)

print("Stock targets shape:", stock_targets.shape)

for horizon in TARGET_HORIZONS:
    print(f"\nFuture return {horizon}d available:")
    display(stock_targets[f"target_available_{horizon}d"].value_counts())

display(
    stock_targets[
        [
            "date",
            "symbol",
            "yf_ticker",
            "adj_close",
            "future_adj_close_5d",
            "future_return_5d",
            "future_adj_close_20d",
            "future_return_20d",
            "future_adj_close_60d",
            "future_return_60d",
        ]
    ]
    .dropna()
    .sample(30, random_state=42)
)
```

### Future downside / upside path targets

Now we create future downside and upside targets.

These are more useful for MarketGuard than only final return.

future_return_20d = where price ended after 20 days.

future_min_return_20d = worst drop seen anytime during next 20 days.

future_max_return_20d = best upside seen anytime during next 20 days.

```python
# Rebuild stock_targets cleanly from stock_features
# This avoids using the partially broken result from the previous failed cell.
stock_targets = stock_features.copy()

# Recreate future adjusted-close prices and future returns
for horizon in TARGET_HORIZONS:
    future_price_col = f"future_adj_close_{horizon}d"
    future_return_col = f"future_return_{horizon}d"
    target_available_col = f"target_available_{horizon}d"

    stock_targets[future_price_col] = (
        stock_targets
        .groupby("yf_ticker")["adj_close"]
        .shift(-horizon)
    )

    stock_targets[future_return_col] = (
        stock_targets[future_price_col] / stock_targets["adj_close"] - 1
    )

    stock_targets[target_available_col] = (
        stock_targets[future_return_col].notna()
    ).astype(int)


def add_future_path_targets(g: pd.DataFrame, horizons: list[int]) -> pd.DataFrame:
    """
    Add future path-based targets for one stock.

    For each date T and horizon H:
    - future_min_return_Hd = lowest adjusted close from T+1 to T+H, relative to T
    - future_max_return_Hd = highest adjusted close from T+1 to T+H, relative to T
    """
    g = g.sort_values("date").copy()

    for horizon in horizons:
        future_price_cols = []

        for step in range(1, horizon + 1):
            col = f"_future_adj_close_step_{step}"
            g[col] = g["adj_close"].shift(-step)
            future_price_cols.append(col)

        full_future_window_available = g[future_price_cols].notna().all(axis=1)

        min_price_col = f"future_min_adj_close_{horizon}d"
        max_price_col = f"future_max_adj_close_{horizon}d"
        min_return_col = f"future_min_return_{horizon}d"
        max_return_col = f"future_max_return_{horizon}d"
        path_available_col = f"path_target_available_{horizon}d"

        g[min_price_col] = g[future_price_cols].min(axis=1)
        g[max_price_col] = g[future_price_cols].max(axis=1)

        g[min_return_col] = g[min_price_col] / g["adj_close"] - 1
        g[max_return_col] = g[max_price_col] / g["adj_close"] - 1

        g.loc[
            ~full_future_window_available,
            [min_price_col, max_price_col, min_return_col, max_return_col]
        ] = np.nan

        g[path_available_col] = full_future_window_available.astype(int)

        g = g.drop(columns=future_price_cols)

    return g


target_frames = []

for ticker, g in stock_targets.groupby("yf_ticker", sort=False):
    gf = add_future_path_targets(g, TARGET_HORIZONS)
    target_frames.append(gf)

stock_targets = (
    pd.concat(target_frames, ignore_index=True)
    .sort_values(["yf_ticker", "date"])
    .reset_index(drop=True)
)

print("Stock targets shape:", stock_targets.shape)
print("yf_ticker exists:", "yf_ticker" in stock_targets.columns)

for horizon in TARGET_HORIZONS:
    print(f"\nFuture path target {horizon}d available:")
    display(stock_targets[f"path_target_available_{horizon}d"].value_counts())

display(
    stock_targets[
        [
            "date",
            "symbol",
            "yf_ticker",
            "adj_close",
            "future_return_20d",
            "future_min_return_20d",
            "future_max_return_20d",
            "target_available_20d",
            "path_target_available_20d",
        ]
    ]
    .dropna()
    .sample(30, random_state=42)
)
```

### 20-day target labels

```python
target_horizon = MAIN_TARGET_HORIZON

target_available_mask = (
    (stock_targets[f"target_available_{target_horizon}d"] == 1)
    & (stock_targets[f"path_target_available_{target_horizon}d"] == 1)
)

# Rows usable for target modelling:
# Features must be ready AND the future 20-day target must be available.
stock_targets["target_ready_v1_20d"] = (
    (stock_targets["model_ready_v1"] == 1)
    & target_available_mask
).astype(int)

stock_targets["target_ready_v1_clean_20d"] = (
    (stock_targets["model_ready_v1_clean"] == 1)
    & target_available_mask
).astype(int)


def make_target_flag(condition: pd.Series, available_mask: pd.Series) -> pd.Series:
    """
    Create nullable target flag.
    - 1 / 0 when target is available
    - NA when future target is not available
    """
    out = pd.Series(pd.NA, index=condition.index, dtype="Int64")
    out.loc[available_mask] = condition.loc[available_mask].astype(int)
    return out


# Final 20-day return direction:
# 1 means stock ended positive after 20 trading days.
stock_targets["target_positive_return_20d"] = make_target_flag(
    stock_targets["future_return_20d"] > 0,
    target_available_mask
)

# Downside risk label:
# 1 means stock fell at least 5% at some point within the next 20 trading days.
stock_targets["target_downside_5pct_20d"] = make_target_flag(
    stock_targets["future_min_return_20d"] <= DOWNSIDE_RISK_THRESHOLD_20D,
    target_available_mask
)

# Big downside risk label:
# 1 means stock fell at least 10% at some point within the next 20 trading days.
stock_targets["target_big_downside_10pct_20d"] = make_target_flag(
    stock_targets["future_min_return_20d"] <= BIG_DOWNSIDE_THRESHOLD_20D,
    target_available_mask
)

# Upside opportunity label:
# 1 means stock rose at least 5% at some point within the next 20 trading days.
stock_targets["target_upside_5pct_20d"] = make_target_flag(
    stock_targets["future_max_return_20d"] >= UPSIDE_THRESHOLD_20D,
    target_available_mask
)

# Simple 3-class forward return bucket.
# This uses final 20-day return, not path high/low.
# -1 = weak, 0 = neutral, 1 = strong
stock_targets["target_return_bucket_20d"] = pd.Series(
    pd.NA,
    index=stock_targets.index,
    dtype="Int64"
)

stock_targets.loc[
    target_available_mask & (stock_targets["future_return_20d"] <= -0.03),
    "target_return_bucket_20d"
] = -1

stock_targets.loc[
    target_available_mask
    & (stock_targets["future_return_20d"] > -0.03)
    & (stock_targets["future_return_20d"] < 0.03),
    "target_return_bucket_20d"
] = 0

stock_targets.loc[
    target_available_mask & (stock_targets["future_return_20d"] >= 0.03),
    "target_return_bucket_20d"
] = 1


print("Stock targets shape:", stock_targets.shape)

print("\nTarget-ready v1 20d:")
display(stock_targets["target_ready_v1_20d"].value_counts())

print("\nTarget-ready v1 clean 20d:")
display(stock_targets["target_ready_v1_clean_20d"].value_counts())

target_label_cols = [
    "target_positive_return_20d",
    "target_downside_5pct_20d",
    "target_big_downside_10pct_20d",
    "target_upside_5pct_20d",
    "target_return_bucket_20d",
]

for col in target_label_cols:
    print(f"\n{col}:")
    display(stock_targets[col].value_counts(dropna=False))

display(
    stock_targets[
        [
            "date",
            "symbol",
            "future_return_20d",
            "future_min_return_20d",
            "future_max_return_20d",
            "target_positive_return_20d",
            "target_downside_5pct_20d",
            "target_big_downside_10pct_20d",
            "target_upside_5pct_20d",
            "target_return_bucket_20d",
            "target_ready_v1_20d",
        ]
    ]
    .dropna()
    .sample(30, random_state=42)
)
```

Now create a more MarketGuard-style scenario target.

This combines future downside and future upside.

### 20-day risk/reward scenario labels

The target asks two yes-or-no questions about the stock’s price movement during the next 20 trading days:

1. Did the stock fall by at least 10%?
2. Did the stock rise by at least 10%?

These two questions create four possible scenarios:

| ID | Hit downside threshold? | Hit upside threshold? | Scenario |
|---:|:---:|:---:|---|
| 0 | No | No | `quiet_or_rangebound` |
| 1 | No | Yes | `upside_low_drawdown` |
| 2 | Yes | Yes | `volatile_two_sided` |
| 3 | Yes | No | `downside_risk_limited_upside` |

```python
# 20-day risk/reward scenario target.
#
# This uses future path behavior:
# - future_min_return_20d = worst drawdown seen within next 20 trading days
# - future_max_return_20d = best upside seen within next 20 trading days
#
# Scenario meanings:
# 0 = quiet_or_rangebound
# 1 = upside_low_drawdown
# 2 = volatile_two_sided
# 3 = downside_risk_limited_upside

scenario_available_mask = target_available_mask.copy()

stock_targets["target_scenario_20d"] = pd.Series(
    pd.NA,
    index=stock_targets.index,
    dtype="string"
)

stock_targets["target_scenario_id_20d"] = pd.Series(
    pd.NA,
    index=stock_targets.index,
    dtype="Int64"
)

future_min_20d = stock_targets["future_min_return_20d"]
future_max_20d = stock_targets["future_max_return_20d"]

quiet_or_rangebound_mask = (
    scenario_available_mask
    & (future_min_20d > DOWNSIDE_RISK_THRESHOLD_20D)
    & (future_max_20d < UPSIDE_THRESHOLD_20D)
)

upside_low_drawdown_mask = (
    scenario_available_mask
    & (future_min_20d > DOWNSIDE_RISK_THRESHOLD_20D)
    & (future_max_20d >= UPSIDE_THRESHOLD_20D)
)

volatile_two_sided_mask = (
    scenario_available_mask
    & (future_min_20d <= DOWNSIDE_RISK_THRESHOLD_20D)
    & (future_max_20d >= UPSIDE_THRESHOLD_20D)
)

downside_risk_limited_upside_mask = (
    scenario_available_mask
    & (future_min_20d <= DOWNSIDE_RISK_THRESHOLD_20D)
    & (future_max_20d < UPSIDE_THRESHOLD_20D)
)

stock_targets.loc[quiet_or_rangebound_mask, "target_scenario_20d"] = "quiet_or_rangebound"
stock_targets.loc[quiet_or_rangebound_mask, "target_scenario_id_20d"] = 0

stock_targets.loc[upside_low_drawdown_mask, "target_scenario_20d"] = "upside_low_drawdown"
stock_targets.loc[upside_low_drawdown_mask, "target_scenario_id_20d"] = 1

stock_targets.loc[volatile_two_sided_mask, "target_scenario_20d"] = "volatile_two_sided"
stock_targets.loc[volatile_two_sided_mask, "target_scenario_id_20d"] = 2

stock_targets.loc[downside_risk_limited_upside_mask, "target_scenario_20d"] = "downside_risk_limited_upside"
stock_targets.loc[downside_risk_limited_upside_mask, "target_scenario_id_20d"] = 3

# Realized future upside/downside ratio.
# This is a target-side diagnostic, not a feature.
future_downside_abs_20d = future_min_20d.clip(upper=0).abs()
future_upside_pos_20d = future_max_20d.clip(lower=0)

stock_targets["future_upside_downside_ratio_20d"] = np.where(
    scenario_available_mask & (future_downside_abs_20d > 0),
    future_upside_pos_20d / future_downside_abs_20d,
    np.nan
)

print("Stock targets shape:", stock_targets.shape)

print("\nScenario distribution, all available 20d targets:")
display(stock_targets["target_scenario_20d"].value_counts(dropna=False))

print("\nScenario distribution inside target_ready_v1_20d:")
display(
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        "target_scenario_20d"
    ].value_counts(dropna=False)
)

print("\nScenario ID distribution inside target_ready_v1_20d:")
display(
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        "target_scenario_id_20d"
    ].value_counts(dropna=False).sort_index()
)

display(
    stock_targets[
        [
            "date",
            "symbol",
            "future_return_20d",
            "future_min_return_20d",
            "future_max_return_20d",
            "future_upside_downside_ratio_20d",
            "target_scenario_20d",
            "target_scenario_id_20d",
            "target_ready_v1_20d",
        ]
    ]
    .dropna()
    .sample(30, random_state=42)
)
```

### Target-ready flags for all horizons

```python
for horizon in TARGET_HORIZONS:
    target_ready_col = f"target_ready_v1_{horizon}d"
    target_ready_clean_col = f"target_ready_v1_clean_{horizon}d"

    available_mask = (
        (stock_targets[f"target_available_{horizon}d"] == 1)
        & (stock_targets[f"path_target_available_{horizon}d"] == 1)
    )

    stock_targets[target_ready_col] = (
        (stock_targets["model_ready_v1"] == 1)
        & available_mask
    ).astype(int)

    stock_targets[target_ready_clean_col] = (
        (stock_targets["model_ready_v1_clean"] == 1)
        & available_mask
    ).astype(int)

print("Stock targets shape:", stock_targets.shape)

for horizon in TARGET_HORIZONS:
    print(f"\nTarget-ready v1 {horizon}d:")
    display(stock_targets[f"target_ready_v1_{horizon}d"].value_counts())

    print(f"\nTarget-ready v1 clean {horizon}d:")
    display(stock_targets[f"target_ready_v1_clean_{horizon}d"].value_counts())
```

### Future relative-return targets

Did the stock beat Nifty 50 over the next 20 days?

Did the stock beat its sector over the next 20 days?

```python
# Future relative-return targets:
# These are target-side values, not input features.
#
# Example:
# future_excess_return_20d_vs_nifty50 =
#     stock future 20d return - Nifty 50 future 20d return

for horizon in TARGET_HORIZONS:
    # Future Nifty 50 return from date T to T+H
    stock_targets[f"future_nifty50_close_{horizon}d"] = (
        stock_targets
        .groupby("yf_ticker")["nifty50_close"]
        .shift(-horizon)
    )

    stock_targets[f"future_nifty50_return_{horizon}d"] = (
        stock_targets[f"future_nifty50_close_{horizon}d"]
        / stock_targets["nifty50_close"]
        - 1
    )

    stock_targets[f"future_excess_return_{horizon}d_vs_nifty50"] = (
        stock_targets[f"future_return_{horizon}d"]
        - stock_targets[f"future_nifty50_return_{horizon}d"]
    )

    # Future sector proxy return from date T to T+H
    stock_targets[f"future_sector_close_{horizon}d"] = (
        stock_targets
        .groupby("yf_ticker")["sector_close"]
        .shift(-horizon)
    )

    stock_targets[f"future_sector_return_{horizon}d"] = (
        stock_targets[f"future_sector_close_{horizon}d"]
        / stock_targets["sector_close"]
        - 1
    )

    stock_targets[f"future_excess_return_{horizon}d_vs_sector"] = (
        stock_targets[f"future_return_{horizon}d"]
        - stock_targets[f"future_sector_return_{horizon}d"]
    )

# 20-day outperform labels
available_20d_mask = stock_targets["target_ready_v1_20d"] == 1

stock_targets["target_outperform_nifty50_20d"] = make_target_flag(
    stock_targets["future_excess_return_20d_vs_nifty50"] > 0,
    available_20d_mask
)

stock_targets["target_outperform_sector_20d"] = make_target_flag(
    stock_targets["future_excess_return_20d_vs_sector"] > 0,
    available_20d_mask
)

print("Stock targets shape:", stock_targets.shape)

print("\n20d outperform Nifty 50:")
display(stock_targets["target_outperform_nifty50_20d"].value_counts(dropna=False))

print("\n20d outperform sector:")
display(stock_targets["target_outperform_sector_20d"].value_counts(dropna=False))

print("\n20d outperform Nifty 50 inside target_ready_v1_20d:")
display(
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        "target_outperform_nifty50_20d"
    ].value_counts(dropna=False)
)

print("\n20d outperform sector inside target_ready_v1_20d:")
display(
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        "target_outperform_sector_20d"
    ].value_counts(dropna=False)
)

display(
    stock_targets[
        [
            "date",
            "symbol",
            "future_return_20d",
            "future_nifty50_return_20d",
            "future_sector_return_20d",
            "future_excess_return_20d_vs_nifty50",
            "future_excess_return_20d_vs_sector",
            "target_outperform_nifty50_20d",
            "target_outperform_sector_20d",
            "target_ready_v1_20d",
        ]
    ]
    .dropna()
    .sample(30, random_state=42)
)
```

### Target quality summary

```python
target_columns = [
    # Future prices / returns
    "future_adj_close_5d",
    "future_return_5d",
    "future_min_return_5d",
    "future_max_return_5d",
    "target_available_5d",
    "path_target_available_5d",
    "target_ready_v1_5d",
    "target_ready_v1_clean_5d",

    "future_adj_close_20d",
    "future_return_20d",
    "future_min_return_20d",
    "future_max_return_20d",
    "target_available_20d",
    "path_target_available_20d",
    "target_ready_v1_20d",
    "target_ready_v1_clean_20d",

    "future_adj_close_60d",
    "future_return_60d",
    "future_min_return_60d",
    "future_max_return_60d",
    "target_available_60d",
    "path_target_available_60d",
    "target_ready_v1_60d",
    "target_ready_v1_clean_60d",

    # 20d labels
    "target_positive_return_20d",
    "target_downside_5pct_20d",
    "target_big_downside_10pct_20d",
    "target_upside_5pct_20d",
    "target_return_bucket_20d",
    "target_scenario_20d",
    "target_scenario_id_20d",
    "future_upside_downside_ratio_20d",

    # Relative future targets
    "future_nifty50_return_20d",
    "future_sector_return_20d",
    "future_excess_return_20d_vs_nifty50",
    "future_excess_return_20d_vs_sector",
    "target_outperform_nifty50_20d",
    "target_outperform_sector_20d",
]

missing_target_cols = [
    col for col in target_columns
    if col not in stock_targets.columns
]

print("Missing target columns:", missing_target_cols)

if missing_target_cols:
    raise ValueError(f"Missing target columns: {missing_target_cols}")

target_quality = []

for col in target_columns:
    target_quality.append(
        {
            "column": col,
            "dtype": str(stock_targets[col].dtype),
            "missing_count": stock_targets[col].isna().sum(),
            "missing_pct": stock_targets[col].isna().mean() * 100,
            "unique_count": stock_targets[col].nunique(dropna=True),
        }
    )

target_quality = (
    pd.DataFrame(target_quality)
    .sort_values("missing_pct", ascending=False)
    .reset_index(drop=True)
)

display(target_quality)

target_quality.to_csv(TARGET_QUALITY_PATH, index=False)

print("Saved:", TARGET_QUALITY_PATH)

print("\nTarget-ready 20d row audit:")
print("Rows:", len(stock_targets))
print("target_ready_v1_20d:", stock_targets["target_ready_v1_20d"].sum())
print("target_ready_v1_clean_20d:", stock_targets["target_ready_v1_clean_20d"].sum())

print("\nMissing labels inside target_ready_v1_20d rows:")
label_cols_for_audit = [
    "target_positive_return_20d",
    "target_downside_5pct_20d",
    "target_big_downside_10pct_20d",
    "target_upside_5pct_20d",
    "target_return_bucket_20d",
    "target_scenario_20d",
    "target_scenario_id_20d",
    "target_outperform_nifty50_20d",
    "target_outperform_sector_20d",
]

display(
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        label_cols_for_audit
    ]
    .isna()
    .sum()
    .reset_index()
    .rename(columns={"index": "target_label", 0: "missing_count"})
)
```

### Final target table audit

```python
# Final target dataset audit before saving

key_cols = ["yf_ticker", "date"]

duplicate_key_rows = stock_targets.duplicated(key_cols).sum()

numeric_cols = stock_targets.select_dtypes(include=["number"]).columns.tolist()

inf_summary = []

for col in numeric_cols:
    s = pd.to_numeric(stock_targets[col], errors="coerce")
    inf_count = np.isinf(s).sum()

    if inf_count > 0:
        inf_summary.append(
            {
                "column": col,
                "inf_count": inf_count,
            }
        )

inf_summary = pd.DataFrame(inf_summary)

# Check target availability consistency
availability_checks = []

for horizon in TARGET_HORIZONS:
    availability_checks.append(
        {
            "horizon": horizon,
            "target_available_count": stock_targets[f"target_available_{horizon}d"].sum(),
            "path_target_available_count": stock_targets[f"path_target_available_{horizon}d"].sum(),
            "target_ready_count": stock_targets[f"target_ready_v1_{horizon}d"].sum(),
            "target_ready_clean_count": stock_targets[f"target_ready_v1_clean_{horizon}d"].sum(),
        }
    )

availability_checks = pd.DataFrame(availability_checks)

# Check missing labels inside 20d target-ready rows
label_cols_for_audit = [
    "target_positive_return_20d",
    "target_downside_5pct_20d",
    "target_big_downside_10pct_20d",
    "target_upside_5pct_20d",
    "target_return_bucket_20d",
    "target_scenario_20d",
    "target_scenario_id_20d",
    "target_outperform_nifty50_20d",
    "target_outperform_sector_20d",
]

missing_labels_inside_ready = (
    stock_targets.loc[
        stock_targets["target_ready_v1_20d"] == 1,
        label_cols_for_audit
    ]
    .isna()
    .sum()
    .reset_index()
)

missing_labels_inside_ready.columns = ["target_label", "missing_count"]

print("Final target dataset shape:", stock_targets.shape)
print("Duplicate yf_ticker-date rows:", duplicate_key_rows)
print("Numeric columns checked:", len(numeric_cols))
print("Columns with infinite values:", len(inf_summary))
print("Latest date:", stock_targets["date"].max())
print("Unique stocks:", stock_targets["yf_ticker"].nunique())

print("\nAvailability checks:")
display(availability_checks)

print("\nInfinite value summary:")
display(inf_summary)

print("\nMissing labels inside target_ready_v1_20d rows:")
display(missing_labels_inside_ready)
```
